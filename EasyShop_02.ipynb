```python
import pandas as pd
import mysql.connector
from textblob import TextBlob

DB_HOST = "localhost"
DB_NAME = "shopeasy"
DB_USER = "root"
DB_PASSWORD = "password12*"

conn = mysql.connector.connect(
    host=DB_HOST,
    database=DB_NAME,
    user=DB_USER,
    password=DB_PASSWORD
)
cursor = conn.cursor()

alter_query = """
    ALTER TABLE customer_reviews 
    ADD COLUMN IF NOT EXISTS polarity FLOAT, 
    ADD COLUMN IF NOT EXISTS subjectivity FLOAT;
"""
try:
    cursor.execute(alter_query)
except:
    pass  

query = "SELECT ReviewID, ReviewText FROM customer_reviews"
cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data, columns=["ReviewID", "ReviewText"])

def analyze_sentiment(text):
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return 0.0, 0.0  # Neutral sentiment for missing or invalid text
    analysis = TextBlob(text)
    return float(analysis.polarity), float(analysis.subjectivity)

df["Polarity"], df["Subjectivity"] = zip(*df["ReviewText"].apply(analyze_sentiment))

update_query = """
    UPDATE customer_reviews
    SET polarity = %s, subjectivity = %s
    WHERE ReviewID = %s
"""
cursor.executemany(update_query, df[["Polarity", "Subjectivity", "ReviewID"]].values.tolist())
conn.commit()
cursor.close()
conn.close()
print("Sentiment analysis results saved to MySQL database.")
```
